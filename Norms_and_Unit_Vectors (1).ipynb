soal no 2

```python
# buat vector v dengan elemen (3,4)
v = np.array([[20, 3, 7, 6], [17, 6, 7, 3], [13, 10, 5, 4]])
v
```

```python
# hitung nilai L2 norm dari vector v
(20**2 + 3**2 + 7**2 + 6**2)**(1/2)
(17**2 + 6**2 + 7**2 + 3**2)**(1/2)
(13**2 + 10**2 + 5**2 + 4**2)**(1/2)
```

```python
# tampilkan unit vector dari vector v
v = np.array([[20, 3, 7, 6], [17, 6, 7, 3], [13, 10, 5, 4]])
print("Unit vector:", v )
```

```python
# hitung panjang unit vector dari vector v
v = np.array([[20, 3, 7, 6], [17, 6, 7, 3], [13, 10, 5, 4]])
print("Panjang vektor:", np.linalg.norm(v))
```
